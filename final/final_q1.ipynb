```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Model parameters
beta = 0.99
kappa = 0.01
phi_pi = 2.5
sigma_n = 0.01
sigma_i = 0.001
```

# (a)

```python
def y_hat(epsil_n, epsil_i):
    return 1/(1+phi_pi * kappa) * epsil_n - 1/(1+phi_pi * kappa) * epsil_i


def pi_hat(epsil_n, epsil_i):
    return kappa/(1+phi_pi * kappa) * epsil_n - kappa/(1+phi_pi * kappa) * epsil_i


def i_hat(epsil_n, epsil_i):
    return (phi_pi * kappa)/(1+phi_pi * kappa) * epsil_n + 1/(1+phi_pi * kappa) * epsil_i

```

```python
np.random.seed(2)
epsilon_n = np.random.normal(0, sigma_n**2, 20)
epsilon_i = np.random.normal(0, sigma_i**2, 20)
epsilon_n[0] = sigma_n

y = y_hat(epsilon_n, epsilon_i)
pi = pi_hat(epsilon_n, epsilon_i)
i = i_hat(epsilon_n, epsilon_i)
ibar = epsilon_i
rn = epsilon_n
r = i

# Plotting
fig, ax = plt.subplots(2, 3, figsize=(20, 10))

ax[0,0].plot(ibar, label=r'$\overline{i}$')
ax[0,0].set_title(r'$\overline{i}$')
ax[0,0].legend()
ax[0,0].grid(True)

ax[0,1].plot(y, label=r'$\hat{y}_t$')
ax[0,1].set_title(r'$\hat{y}_t$')
ax[0,1].legend()
ax[0,1].grid(True)

ax[0,2].plot(pi, label=r'$\hat{pi}_t$')
ax[0,2].set_title(r'$\hat{pi}_t$')
ax[0,2].legend()
ax[0,2].grid(True)

ax[1,0].plot(i, label=r'$\hat{i}_t$')
ax[1,0].set_title(r'$\hat{i}_t$')
ax[1,0].legend()
ax[1,0].grid(True)

ax[1,1].plot(r, label=r'$\hat{r}_t$')
ax[1,1].set_title(r'$\hat{r}_t$')
ax[1,1].legend()
ax[1,1].grid(True)

ax[1,2].plot(rn, label=r'$\hat{r}^{n}_t$')
ax[1,2].set_title(r'$\hat{r}^{n}_t$')
ax[1,2].legend()
ax[1,2].grid(True)

plt.savefig("q1a_nshcok.png", dpi=300)
```
